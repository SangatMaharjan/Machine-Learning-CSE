```python
#Working with Dictionary

# A data structure used to store data in key value pairs

student= {
    "name" : "John" ,
    "age" : 20 ,
    "faculty" : "BCE"
}

print(student["name"])
```

```python
# Access Value

print(student["name"])
```

```python
# Use Get() function

student.get("name")
```

```python
# Add item

student["address"] = "Lalitpur"

print(student)
```

```python
# Update Value

student["name"]= "Jacob"
print(student)
```

```python
# Remove item using del

del student["age"]
print(student)
```

```python
# Remove using pop

student.pop("faculty")
print(student)
```

```python
# Remove Last Item

student.popitem()
print(student)
```

```python
# Clear Dictionary

student.clear()
print(student)
```

```python
# Copy Dictionary

student = {
    "name": "John",
    "age": 20,
    "faculty": "BCE"
}

# Use an underscore instead of a space
new_stud = student.copy()

print(new_stud)
```

```python
# Get All Keys
new_stud.keys()

```

```python
# Get All Values
new_stud.values()
```

```python
# Get All Items

new_stud.items()
```

```python
# Check Key Exists

"name" in new_stud
"marks" in new_stud
"address" in new_stud
```

```python
# Get Length

len(new_stud)
```

```python
# Loop Through Keys

for k in new_stud:
  print(k)
```

```python
#Loop Through Values

for v in new_stud.values():
  print(v)
```

```python
# Loop Through Items

for k, v in new_stud.items():
    print(f"key: {k} and Value: {v}")
```

```python
# Working With Pandas DataFrame

''' A DataFrame in Python is a two-dimensional, tabular data structure that organizes
data into labeled rows and columns. It is conceptually identical to an Excel
spreadsheet or a SQL table, making it the primary tool for data manipulation and
analysis in the Pandas Library. '''

```

```python
# Create a DataFrame

import pandas as pd
student1= {
    "name" : "John" ,
    "age" : 20 ,
    "faculty" : "BCE" ,
    "marks":91
}

student2={
    "name" : "Jacob" ,
    "age" : 21,
    "faculty" : "BCE" ,
    "marks":68
}

student3= {
    "name" : "Rachel" ,
    "age": 23 ,
    "faculty":"BCV" ,
    "marks" : 81
}
student4={
    "name": "Roman" ,
    "age":20 ,
    "faculty":"ARCH" ,
    "marks" : 78
}
student5= {
    "name": "Rock" ,
    "age": 19 ,
    "faculty":"BEIT" ,
    "marks":89
}

student6={
    "name": "Victor" ,
    "age": 21 ,
    "faculty": "BCE" ,
    "marks": 97
}
df = pd.DataFrame([student6 , student1, student2,student3,student4 , student5])
print(df)

df1=pd.DataFrame([student1, student2, student3])


```

```python
#Add a new column called Grade to the DataFrame

df['Grade']=['A','A','C','B+','B','A-']
print(df)


```

```python
# Increase all students marks by 5

df['marks'] = df['marks']+ 5
```

```python
# Update the Marks of Jacob to 85 using a condition

df.loc[df['name']=='Jacob','marks']=95
print(df)
```

```python
def assign_grade(mark):
    if mark >= 90:
        return 'A+'
    elif mark >= 80:
        return 'A'
    elif mark >= 70:
        return 'B'
    else:
        return 'F'

df['grade'] = df['marks'].apply(assign_grade)
print(df)
```

```python
# Remove the other Grade Column

# axis =1 signifies we remove the entire vertical column
df=df.drop('Grade',axis=1)
print(df)
```

```python
# Remove the first row from the data frame

df=df.drop(0, axis=0)
print(df)
```

```python
#Sort the DataFrame by Marks in ascending order

df.sort_values('marks')
```

```python
# Sort the DataFramy by marks in Descending Order

df.sort_values('marks', ascending=False)
```
